```python
import os
from tqdm import tqdm
from PyPDF2 import PdfFileReader
```

```python
path = "E:/"
```

```python
import docx
def read_word_file(path):
    doc = docx.Document(path)
    text = []
    for paragraph in doc.paragraphs:
        text.append(paragraph.text)
    return '\n'.join(text)
```

```python
def searchPC(path,searchingTerm, searchPDFs=True, searchWordFiles=True, searchTextFiles=True):
    '''Path is like E:/ for example'''
    foundList = []
    searchingTerm = searchingTerm.lower()
    for root, dirs, files in tqdm(os.walk(path)):
        for f in files:
            #print(os.path.join(root,f))
            if(searchPDFs):
                try:
                    if(f[-3].lower() == 'p' and f[-2].lower() == 'd' and f[-1].lower() == 'f'):
                        try:
                            pdf = PdfFileReader(os.path.join(root,f))
                        except:
                            break
                        for pageNo in range(0, pdf.numPages):
                            pageObj = pdf.getPage(pageNo)
                            try:
                                txt = (pageObj.extractText()).lower
                                #print(txt)
                                if(searchingTerm in txt):
                                    if(os.path.join(root,f) not in foundList):
                                        foundList.append(os.path.join(root,f))
                            except:
                                uselessFlag = True
                except:
                    uselessFlag = True
            if(searchTextFiles):
                try:
                    if(f[-3].lower() == 't' and f[-2].lower() == 'x' and f[-1].lower() == 't'):
                        try:
                            txtFile = open(os.path.join(root,f), "r")
                            #print(os.path.join(root,f))
                            txt = (txtFile.read()).lower()
                            if(searchingTerm in txt):
                                if(os.path.join(root,f) not in foundList):
                                    foundList.append(os.path.join(root,f))
                        except:
                            uselessFlag = True
                except:
                    uselessFlag = True
            if(searchWordFiles):
                try:
                    if(f[-4].lower() == 'd' and f[-3].lower() == 'o' and f[-2].lower() == 'c' and f[-1].lower() == 'x'):
                        txt = (read_word_file(os.path.join(root,f))).lower()
                        if(searchingTerm in txt):
                            if(os.path.join(root,f) not in foundList):
                                foundList.append(os.path.join(root,f))
                except:
                    uselessFlag = True
    return foundList
            
```

```python
founded = searchPC(path, "Data Structure")
```

```python
for i in founded:
    print(i)
```
